```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
education_districtwise = pd.read_csv('/kaggle/input/education-districtwise-file/education_districtwise.csv')
```

```python
sampled_data = education_districtwise.sample(n=50, replace=True, random_state=31208)
sampled_data
```

The output show 50 districts that have been selected randomly from the dataset. Note that each has a different literacy rate.

95% Confidence Interval for the Mean District literacy rate base on the above sample data.

```python
sample_mean = sampled_data['OVERALL_LI'].mean()
```

```python
estimated_standard_error = sampled_data['OVERALL_LI'].std() / np.sqrt(sampled_data.shape[0])
```

```python
stats.norm.interval(confidence=0.95, loc=sample_mean, scale=estimated_standard_error)
```

We have a 95% confidence interval for the mean district literacy rate that stretches from about 71.7% to 77%

95% Cl(71.73, 77.02). Thus the Department of Education will use the estimate of the mean district literacy rate to help make decisions about distribution of funds to different states.

99% Confidence Interval

```python
stats.norm.interval(confidence=0.99, loc=sample_mean, scale=estimated_standard_error)
```

There is a 99% confidence interval for the mean district literacy rate that stretches from about 70.9% to 77.9%.
99% Cl:(70.90,77.90)

The relationship between the CL & CI: As the CL gets higher, ther CI gets wider. ie 

With confidence level of 95%, the interval covers 5.27 percentage points (77%- 71.73%)

With a confidence level of 99%, the interval covers 7 percentage points(77.9%-70.9%)

A wider confidence level interval is more likely toinclude the actual population parameter.
With the above results, Department of Education can decide hpw to distribute government resources to improve literacy.
